```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
size = np.array([800, 1000, 1200, 1500, 1800]).reshape(-1, 1)
price = np.array([150000, 180000, 200000, 250000, 300000])
```

```python
data = pd.DataFrame({'House Size (sq ft)': size.flatten(), 'Price (USD)': price})
print("House Size and Price (USD):")
print(data)
```

```python
model = LinearRegression()
model.fit(size, price)
```

```python
predicted_price = model.predict([[1300]])
predicted_price_inr = predicted_price[0] * 82
```

```python
print(f"\nPredicted price for a 1300 sq ft house: ₹{predicted_price_inr:.2f}")
```

```python
plt.scatter(size, price * 82, color='blue', label='Data points')
plt.plot(size, model.predict(size) * 82, color='red', label='Regression line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (INR)')
plt.title('Simple Linear Regression: House Price Prediction (INR)')
plt.legend()
plt.show()
```
