```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
experience = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
base_salary = 30000
increment_rate = 0.15
salary = np.array([base_salary * ((1 + increment_rate) ** (year - 1)) for year in range(1, 8)])
```

```python

data = pd.DataFrame({'Years of Experience': experience.flatten(), 'Fair Salary (INR)': salary.astype(int)})
print("Years of Experience and Fair Salary (INR):")
print(data)
```

```python
model = LinearRegression()
model.fit(experience, salary)
```

```python
predicted_salary = model.predict([[4.5]])
print(f"\nPredicted fair salary for 4.5 years of experience: ₹{predicted_salary[0]:.2f}")
```

```python
plt.scatter(experience, salary, color='blue', label='Fair Salary Data')
plt.plot(experience, model.predict(experience), color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Fair Salary (INR)')
plt.title('Fair Salary Prediction using Linear Regression')
plt.legend()
plt.show()
```
